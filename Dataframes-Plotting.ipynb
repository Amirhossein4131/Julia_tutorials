# Data Frames

```julia
import Pkg
Pkg.add("DataFrames")
```

```julia
using DataFrames
df = DataFrame(Name = ["Alice", "Bob", "Charlie"], Age = [25, 30, 35], Salary = [50000, 60000, 70000])
```

```julia
# Accessing a column
names = df.Name
println(names)
```

```julia
# Accessing rows
row_1 = df[1, :]
println(row_1)
```

```julia
# Accessing specific element
element = df[2, :Salary]
println(element)
```

#### Adding columns

```julia
# Add column
df.Gender = ["M", "F", "F"]

# Manipulating column
df.Age .+= 1

println(df)
```

#### Data Filtering

```julia
# Salary more than 55000
high_salary = df[df.Salary .> 55000, :]
println(high_salary)
```

#### Sort Column

```julia
# Sorrt by salary in decending order
sorted = sort(df, :Salary, rev=true)
```

```julia
describe(df)
```

# Plotting

```julia
Pkg.add("Plots")
```

```julia
using Plots

x = 1:10
y = rand(10)

# A basic plot
plot(x, y, label="Random", xlabel="X", ylabel="Y", title="Customized Plot",
     color=:red, linewidth=2, marker=:circle, markersize=8)
```

```julia
#### Scatter Plot
scatter(x, y, label="Scatter Plot", xlabel="X", ylabel="Y", title="Scatter Example", color=:blue)
```

```julia
#### Bar plot
bar(x, y, label="Bar Plot", xlabel="X", ylabel="Y", title="Bar Example", color=:green)
```

```julia
#### Histogram
data = randn(1000)  # 1000 random values from a normal distribution
histogram(data, bins=20, xlabel="X", ylabel="Y", label="Histogram", title="Histogram Example", color=:orange, dpi=300)
histogram!( data .+= 2, bins=20, color=:blue, alpha=0.5, label="Histogram2", dpi=300)

#### Save figs
savefig("Example_saved_plot.png")
```

```julia
#### HeatMaps
data = rand(10, 10)
heatmap(data, title="Customized Heatmap", xlabel="X-axis", ylabel="Y-axis", 
        color=:viridis)
```

```julia
x = 1:10
y = 1:10
z = [sin(i) * cos(j) for i in x, j in y]  # 2D array using a mathematical function

heatmap(x, y, z, title="Heatmap with Custom Data", xlabel="X", ylabel="Y", color=:plasma,clims=(-1, 1), dpi=300)

#savefig("heatmap_example.png")
```

```julia
#### Contour Plots 
# Define grid points
x = -5:0.1:5
y = -5:0.1:5

# Define the 2D function to plot
z = [sin(sqrt(xi^2 + yi^2)) for xi in x, yi in y]

# Create the filled contour plot
contourf(x, y, z, title="Filled Contour Plot", xlabel="X-axis", ylabel="Y-axis")
scatter!(x[1:10:end], y[1:10:end], z[1:10:end], color=:white, label="Scatter Points")

#savefig("Contour_plot.png")
```
