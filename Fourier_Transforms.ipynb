## Analyzing Heat Diffusion

#### perform a Fourier Transform to analyze the frequency spectrum of the heat distribution over time.

#### Define the heat equation

$$
\frac{\partial u(x, y, t)}{\partial t} = \alpha \left( \frac{\partial^2 u(x, y, t)}{\partial x^2} + \frac{\partial^2 u(x, y, t)}{\partial y^2} \right)
$$

Where:

u(x,y,t) is the temperature distribution.

α is the thermal diffusivity.

```julia
# Parameters
Nx, Ny = 256, 256  # Grid points in x and y directions
Lx, Ly = 10.0, 10.0  # Length of the domain
α = 0.4  # Thermal diffusivity
dt = 0.01  # Time step
T = 10.0  # Total time unit
Nt = Int(T / dt)  # Number of time steps

# Ensure x and y are sorted in ascending order
x = sort(LinRange(-Lx/2, Lx/2, Nx))  # Sort x points in ascending order
y = sort(LinRange(-Ly/2, Ly/2, Ny))  # Sort y points in ascending order
dx = x[2] - x[1]
dy = y[2] - y[1]
```

#### initial condition as a Gaussian heat distribution
$$
u(x, y, 0) = e^{-10(x^2 + y^2)}
$$

```julia
using FFTW, Plots, LinearAlgebra

# Create initial condition (Gaussian heat spot)
u_initial = [exp(-10 * (x[i]^2 + y[j]^2)) for i in 1:Nx, j in 1:Ny]

# FFT of initial condition
u_hat = fft(u_initial)

heatmap(x, y, u_initial, title="Heat Diffusion at t = 0", color=:inferno)
```

#### Evolve the heat using Fourier Transforms

Time Evolution:
- The loop iterates over each time step, updating the Fourier coefficients u_hat.
- The expression u_hat .* exp.(-α * K2 * dt) implements the heat equation in the frequency domain.
- ifft(u_hat): Converts the updated frequency domain back to the spatial domain to get the temperature distribution at the current time

```julia
# Fourier frequencies (wave numbers)
kx = fftshift(range(-Nx/2, stop=Nx/2-1) * (2 * π / Lx))
ky = fftshift(range(-Ny/2, stop=Ny/2-1) * (2 * π / Ly))

# Creating the mesh grid manually using broadcasting
Kx = reshape(kx, Nx, 1) .* ones(1, Ny)
Ky = ones(Nx, 1) .* reshape(ky, 1, Ny)
K2 = Kx.^2 + Ky.^2  # Wave number squared (|k|^2)

# Time evolution using FFT
for t in 1:Nt
    u_hat .= u_hat .* exp.(-α * K2 * dt)  # Update in frequency space

    # Inverse FFT to get back to spatial domain
    u = real(ifft(u_hat))

    # Visualization every few steps
    if t % 500 == 0
        heatmap(x, y, u, title="Heat Diffusion at t = $(t * dt)", color=:inferno)
        display(current())  # Display the current plot
        sleep(0.1)  # Short delay to allow real-time updates
    end
end
```
