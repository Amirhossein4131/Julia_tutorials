# Neural Nets -- MNIST calssification


#### function for data loading

```julia
using Flux, MLDatasets, Plots, Random
using Flux: params, onehotbatch, crossentropy

# load data
function load_and_preprocess_data()
    train_x, train_y = MNIST(:train)[:]
    test_x, test_y = MNIST(:test)[:]

    train_x = Flux.flatten(train_x) |> x -> x / 255.0
    test_x = Flux.flatten(test_x) |> x -> x / 255.0

    train_y = Flux.onehotbatch(train_y, 0:9)
    test_y = Flux.onehotbatch(test_y, 0:9)

    return (train_x, train_y, test_x, test_y)
end
```

#### Create Model

```julia
function define_model()
    return Chain(
        Dense(28 * 28, 64, relu),  # Input layer
        Dense(64, 10),             # Output layer
        softmax                    # Softmax activation for classification
    )
end
```

#### Create Loss module

```julia
function loss_function()
    loss(x, y) = crossentropy(model(x), y)
    opt = ADAM()
    return loss, opt
end
```

#### Get batches

```julia
function get_batches(x, y, batch_size)
    n = size(x, 2)
    indices = randperm(n)
    batches = [
        (x[:, indices[i:min(i + batch_size - 1, n)]], 
         y[:, indices[i:min(i + batch_size - 1, n)]])
        for i in 1:batch_size:n
    ]
    return batches
end
```

#### train function

```julia
function train_model(model, train_x, train_y, test_x, test_y, epochs=50, batch_size=128)
    loss, opt = loss_function()

    train_loss_history = []
    test_loss_history = []

    for epoch in 1:epochs
        batches = get_batches(train_x, train_y, batch_size)

        for (x_batch, y_batch) in batches
            Flux.train!(loss, params(model), [(x_batch, y_batch)], opt)
        end

        epoch_train_loss = loss(train_x, train_y)
        epoch_test_loss = loss(test_x, test_y)

        push!(train_loss_history, epoch_train_loss)
        push!(test_loss_history, epoch_test_loss)

        println("Epoch $epoch, Train loss: $epoch_train_loss, Test loss: $epoch_test_loss")
    end

    return train_loss_history, test_loss_history
end
```

#### train

```julia
(train_x, train_y, test_x, test_y) = load_and_preprocess_data()
model = define_model()
train_loss_history, test_loss_history = train_model(model, train_x, train_y, test_x, test_y)
```

#### Plot

```julia
plot(1:length(train_loss_history), train_loss_history, marker=:o, label="Train Loss", title="Training and Test Loss", xlabel="Epoch", ylabel="Loss", lw=2)
plot!(1:length(test_loss_history), test_loss_history, marker=:x, label="Test Loss", lw=2)
```
