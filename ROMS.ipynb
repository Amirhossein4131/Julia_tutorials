### Time Evolution of a 1D heat transfer:

- The temperature distribution is updated using a finite difference scheme based on the heat equation:

$$
u_i^{n+1} = u_i^n + \alpha \frac{\Delta t}{(\Delta x)^2} \left( u_{i+1}^n - 2u_i^n + u_{i-1}^n \right)
$$

where:

- $u_i^n$ is the temperature at spatial point 

- Δt is the time step, and Δx is the spatial step.

```julia
using LinearAlgebra, Plots
using Random
using BenchmarkTools  # For benchmarking

# Step 1: Discretize the 1D heat equation
function heat_equation_simulation(nx::Int, nt::Int, α::Float64, tmax::Float64)
    dx = 1 / (nx - 1)  # Spatial step size
    dt = tmax / nt     # Time step size
    
    x = range(0, stop=1, length=nx)
    u = zeros(nx, nt)  # Temperature distribution, u(x,t)
    
    # Initial condition: Heat at the center
    u[:,1] .= exp.(-100 * (x .- 0.5).^2)
    
    # Loop over time steps
    for t in 2:nt
        for i in 2:nx-1
            u[i,t] = u[i,t-1] + α * dt / dx^2 * (u[i+1,t-1] - 2*u[i,t-1] + u[i-1,t-1])
        end
    end
    
    return u, x
end
```

```julia
# Parameters
nx = 100  # Number of spatial points
nt = 500  # Number of time steps
α = 0.01  # Thermal diffusivity constant
tmax = 2.0  # Total simulation time

# Simulate the heat equation
u, x = heat_equation_simulation(nx, nt, α, tmax)

size(u)
```

## Perform Proper Orthogonal Decomposition (POD)


#### Step 2.1: Compute the covariance matrix

```julia
# Each column of u is a snapshot at a different time step
U = u[:,1:nt]  # Use all snapshots as input
C = U * U' / nt
size(C)
```

##### Step 2.2: Singular Value Decomposition (SVD)

```julia
Φ, Σ, Ψ = svd(C)
println(size(Φ))
println(size(Σ))
println(size(Ψ))
```

##### Step 3: Truncate the SVD

```julia
r = 1  # Number of retained modes (rank reduction)
Φ_r = Φ[:, 1:r]  # Truncated basis functions
size(Φ_r)
```

#### Step 4: Project the original data onto the reduced basis

```julia
u_reduced = Φ_r' * U
size(u_reduced)
```

#### Step 5: Reconstruct the original system using the reduced basis

```julia
u_reconstructed = Φ_r * u_reduced
size(u_reconstructed)
```

### Dimensionality comparison

```julia
original_dimensionality = size(u, 2)  # Original number of time steps
reduced_dimensionality = size(u_reduced, 1)  # Reduced modes
println("Original dimensionality (time steps): ", original_dimensionality)
println("Reduced dimensionality (modes): ", reduced_dimensionality)
```

### Computation time comparison

```julia
# Measure computational time for the original model
println("Computation time for exact solution")
@btime heat_equation_simulation(nx, nt, α, tmax)

println("Computational time for ROM")
# Measure computational time for POD
@btime u_reconstructed = Φ_r * u_reduced

println("The ROM is faster")
```

#### Plot original VS POD

```julia
# Plot the original solution and the reduced-order approximation
plot(x, u[:, 1], label="Original: t=0", xlabel="x", ylabel="u(x,t)", lw=10, legend=:topright, alpha=0.5)
plot!(x, u[:, nt÷4], label="Original: t=0.5tmax", lw=10, alpha=0.5)
plot!(x, u[:, nt÷2], label="Original: t=tmax", lw=10, alpha=0.5)

# Plot the reduced-order approximation at the same time steps
plot!(x, u_reconstructed[:, 1], label="Reduced: t=0", linestyle=:dash, lw=2)
plot!(x, u_reconstructed[:, nt÷4], label="Reduced: t=0.5tmax", linestyle=:dash, lw=2)
plot!(x, u_reconstructed[:, nt÷2], label="Reduced: t=tmax", linestyle=:dash, lw=2)
```

```julia
# Plot the reconstruction error
error_plot = plot(x, abs.(u[:, nt÷2] .- u_reconstructed[:, nt÷2]), 
                  xlabel="x", ylabel="Error", title="Reconstruction Error at t=tmax", lw=2)
display(error_plot)
```

### N.B: The errors go very low if the number of modes are increased.
